# Question 1

```python
import pandas as pd

wine = pd.read_csv("D:\\360DigiTMG\\DataScience\\11. Data mining Unsupervised learning - Dimention Reduction (PCA)\\Assignement\\wine.csv")
wine.head()
```

```python
wine.describe()
```

```python
from sklearn.preprocessing import StandardScaler
wine_scl = StandardScaler()
std_wine = pd.DataFrame(wine_scl.fit_transform(wine), columns=wine.columns)
std_wine
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

before_TWSS = []
before_k = list(range(2,9))
for i in before_k:
    before_wine_kmeans = KMeans(n_clusters=i).fit(std_wine)
    before_TWSS.append(before_wine_kmeans.inertia_)
plt.plot(before_k, before_TWSS)
plt.show()
```

```python
before_wine_kmeans = KMeans(n_clusters=3).fit(std_wine)
before_wine_kmeans_labels = pd.Series(before_wine_kmeans.labels_)
wine['clusters'] = before_wine_kmeans_labels
wine
```

```python
wine[wine['clusters'] == 0].describe()
```

```python
wine[wine['clusters'] == 1].describe()
```

```python
wine[wine['clusters'] == 2].describe()
```

```python
std_wine
```

```python
# Doing PCA with the standardized data
from sklearn.decomposition import PCA

wine_pca = PCA()
wine_pca.fit(std_wine)
wine_pca.explained_variance_ratio_
```

```python
plt.plot(range(1,15), wine_pca.explained_variance_ratio_.cumsum(), marker='o', linestyle="-")
```

```python
wine_pca = PCA(n_components=6)
wine_pca_components = wine_pca.fit_transform(std_wine)
wine_pca_components = pd.DataFrame(wine_pca_components)
wine_pca_components.columns = 'comp0', 'comp1', 'comp2', 'comp3', 'comp4', 'comp5'
wine_pca_components
```

```python
wine_pca_components.plot(x='comp2', y='comp1', kind='scatter')
```

```python
after_TWSS = []
after_k = list(range(2,9))
for i in after_k:
    after_wine_kmeans = KMeans(n_clusters=i).fit(wine_pca_components)
    after_TWSS.append(after_wine_kmeans.inertia_)
plt.plot(after_k, after_TWSS)
plt.show()
```

```python
after_wine_kmeans = KMeans(n_clusters=3).fit(wine_pca_components)
after_wine_kmeans_labels = pd.Series(after_wine_kmeans.labels_)
after_wine_kmeans_labels
```

```python
wine['pca_clusters'] = after_wine_kmeans_labels
wine
```

# Question 2

```python
import pandas as pd
heart_disease = pd.read_csv("D:\\360DigiTMG\\DataScience\\11. Data mining Unsupervised learning - Dimention Reduction (PCA)\\Assignement\\heart disease.csv")
heart_disease.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms_heart_disease = MinMaxScaler()
mms_heart_disease_data = pd.DataFrame(mms_heart_disease.fit_transform(heart_disease), columns=heart_disease.columns)
mms_heart_disease_data.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

before_z = linkage(mms_heart_disease_data, method="complete", metric="euclidean")
dendrogram(before_z, truncate_mode="lastp")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

before_heart_disease_clustering = AgglomerativeClustering(n_clusters=2, linkage="complete", affinity="euclidean").fit(mms_heart_disease_data)
before_heart_disease_clustering.labels_
```

```python
before_heart_disease_clustering_labels = pd.Series(before_heart_disease_clustering.labels_)
before_heart_disease_clustering_labels
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_values = pca.fit(mms_heart_disease_data)
pca_values.explained_variance_ratio_
```

```python
plt.plot(range(1,15), pca_values.explained_variance_ratio_.cumsum(), marker="o", linestyle="--")
```

```python
pca = PCA(n_components=6)
pca_values = pca.fit_transform(mms_heart_disease_data)
pca_values_heart_disease = pd.DataFrame(pca_values)
pca_values_heart_disease.columns = 'comp0', 'comp1', 'comp2', 'comp3', 'comp4', 'comp5'
pca_values_heart_disease
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
after_z = linkage(pca_values_heart_disease, method="complete", metric="euclidean")
dendrogram(after_z, truncate_mode="lastp")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
after_heart_disease_clustering = AgglomerativeClustering(n_clusters=2, linkage="complete", affinity="euclidean").fit(pca_values_heart_disease)
after_heart_disease_clustering_labels = pd.Series(after_heart_disease_clustering.labels_)
after_heart_disease_clustering_labels
```

```python
heart_disease['cluster_before'] = before_heart_disease_clustering_labels
heart_disease['cluster_after'] = after_heart_disease_clustering_labels
heart_disease.tail(50)
```
